### Question6

```python
word = input("Enter word to be translated:\n")

vowels = ['a','e','i','o','u','A','E','I','O','U']
def pig(word): 
    first = word[0]

    if first in vowels: 
        word = word.lower()
        word += "way" 
        return word
    else: 
        word = word.lower()

        for letter in word:
            if letter in vowels:
                index_value = word.index(letter)
                break

        word = word[index_value:] +word[:index_value]+ "ay" 
        return word 

print(pig(word))
```
